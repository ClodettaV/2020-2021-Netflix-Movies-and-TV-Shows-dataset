```python
# Import the Pandas library
import pandas as pd

# Import SQL Alchemy
from sqlalchemy import create_engine
```

```python
# Create a reference the CSV file desired
csv_path = "netflix_movies_tvshows.csv"

# Read the CSV into a Pandas DataFrame
Netflix_df = pd.read_csv(csv_path)

# Print the dataframe
Netflix_df.head()
```

```python
# Create Database Connection
engine = create_engine("sqlite:///Netflix_moviesandTvshows.sqlite")

#Export table as sqlite database
Netflix_df.to_sql("netflix", engine, if_exists="replace", index=False)
```

```python
df=pd.read_sql("select * from netflix",engine)
df
year="2020,2021"
type= "TV Show"
grouped= df[(df["release_year"].isin(map(int,year.split(','))))&(df["type"].isin(type.split(',')))].groupby(["director","type"]).count().sort_values("title",ascending=False)["title"].reset_index()
grouped[0:10]
```
